```python
from pickletools import optimize
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.transforms import transforms
import custom_dataset
import augmentation
```

```python
data_dir = "../landmark_classification/dataset/"
transform = transforms.Compose([
    transforms.ToTensor()
])


dataset = custom_dataset.Dataset(os.path.join(data_dir, 'train'), transform=transform)
```

```python
images = dataset.images
categorized_images = []
for i in range(10):
    categorized_images.append(dataset.categorized_images(i, os.path.join(data_dir, 'train.csv')))
```

```python
def show_images(images, images_per_row=5):
    n_images = len(images)
    n_rows = ((n_images-1) // images_per_row) + 1
    plt.figure(figsize=(20,50))
    for i in range(n_images):
        plt.subplot(n_rows, images_per_row, i+1)
        plt.subplots_adjust(hspace=None, wspace=None)
        plt.imshow(images[i].permute(1, 2, 0))
        plt.axis('off')
    plt.show()
```
